```python
import matplotlib.pyplot as plt

# Create the data lists
years_of_experience = [1, 2, 3, 4, 4.5, 5, 5.5, 6, 7, 8, 9.5]
percentiles = [30, 52, 71, 80, 82, 85, 88, 89, 93, 95, 97]

# Create the plot
plt.plot(years_of_experience, percentiles)

# Set the axis labels
plt.xlabel("Years of Experience")
plt.ylabel("Percentile")

# Set the title
plt.title("Years of Experience vs. Percentile")

# Show the plot
plt.show()
```
